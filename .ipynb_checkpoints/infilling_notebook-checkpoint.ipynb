```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from PIL import Image
from pytorch_metric_learning import losses
from skimage import io, transform
import time
import torch
from torch.autograd import Variable
import torch.nn as nn
from torch.utils.data.sampler import SubsetRandomSampler
from torchvision import datasets, models, transforms
```

```python
# Configurations

# Device
device = torch.device("cuda:2" if torch.cuda.is_available() else "cpu")

# Num epochs
num_epochs = 25

# Model 
model = models.resnet18()

# Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Batch size
batch_size = 32

# Data set
train_path = '/lab/vislab/DATA/CUB/images/'

# Loss function
criterion = losses.TripletMarginLoss(margin=0.1)
```

```python
# Data Loader
transformations = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.ToTensor(),
])
dataset = datasets.ImageFolder(train_path, transformations)

dataset_size = len(dataset)
indices = list(range(dataset_size))
split = int(np.floor(.2 * dataset_size))
np.random.seed(32)
np.random.shuffle(indices)
train_indices, val_indices = indices[split:], indices[:split]

# Creating PT data samplers and loaders:
train_sampler = SubsetRandomSampler(train_indices)
valid_sampler = SubsetRandomSampler(val_indices)

train_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, 
                                           sampler=train_sampler, num_workers=4)
validation_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size,
                                                sampler=valid_sampler)
```

```python
def train_model():
    """generic function to train model"""
    
    optimizer.zero_grad()
    train_loss = []
    loss_values= []
    start_time = datetime.now()
    num_batches = 0

    for epoch in range(num_epochs): 
        print("epoch num:", epoch)
        
        # remove , 'val' to skip validation stage
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()
                
            running_loss = 0.0
            n_dev_correct = 0.0

            for batch_idx, (inputs, labels) in enumerate(train_loader): 
                if phase == 'train':
                    optimizer.zero_grad()

                inputs, labels = inputs.to(device), labels.to(device)
                output = model.forward(inputs)

                loss = criterion(output, labels)
                loss = Variable(loss, requires_grad = True)
                
                if phase == 'train':
                    loss.backward()
                    optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                train_loss.append(loss)
                num_batches += 1



            time_elapsed = datetime.now() - start_time 
            print('Time elapsed (hh:mm:ss.ms) {}'.format(time_elapsed))
            loss_values.append(running_loss / num_batches)
         
        # validate at end of each epoch:
        

    plt.plot(loss_values)    
    
    return model, running_loss
```

```python
# Run Script
model.to(device)

trained_model, loss = train_model()
print(loss)
```
